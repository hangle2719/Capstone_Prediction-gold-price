```python
driver=webdriver.Chrome()
```

```python
driver.get("https://vn.investing.com/commodities/gold-historical-data")
wait = WebDriverWait(driver, 200)
```

```python
date_picker = driver.find_element_by_xpath('//input[@class="hasDatepicker"]')
```

```python
import time
#Set start date - end Date
driver.execute_script("arguments[0].setAttribute('value','30/06/2016 - 30/06/2021')", date_picker)
#find activate calendar widget
date_picker_1 = driver.find_element_by_id('widgetFieldDateRange')
date_picker_1.click()
time.sleep(2)
#Apply new search value
apply_button = driver.find_elements_by_id('applyBtn')
apply_button[0].click()
```
